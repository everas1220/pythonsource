```python
# score 읽어와서 평균,총점 출력

scores = []

f = open("data/score.txt","r", encoding="utf-8")
line = f.readline) # 줄 단위로 읽어오기
while line:    
    scores.append(int(line))
    line = f.readline()
f.close()
```

```python
print(scores)
print("합계 : ",sum(scores))
print("평균 : %.2f" % (sum(scores)/len(scores)))
print("평균 : {:6.3}".format(sum(scores)/len(scores)))
```

```python
# info.txt 읽어와서 BMI 지수 계산후 화면 출력
f = open("data/info.txt","r", encoding="utf-8")
for info in f:
    name, weight, height = info.strip().split(", ")
    print(name,weight,height)
    # BMI = weight / ((height / 100) ** 2)
    bmi = int(weight) / ((int(height) / 100) ** 2)

    result = "";
    # bmi >= 25 : 과체중
    if bmi >= 25:
        result = "과체중"
    # bmi >= 18.5 : 정상체중
    elif bmi >= 18.5:
        result = "정상체중"
    #  : 저체중
    else:
        result = "저체중"

    # 나다 82 180
    # 이름 : 나다
    # print(f"이름 : {name}")
    # 몸무게 : 82
    # 키 : 180
    # bmi : 33.78
    # 결과 : 과체중

    print("\n".join(["이름 : {}", "몸무게 : {}", "키 : {}", "BMI : {}", "결과 : {}"]).format(name, weight, height,bmi,result))
    print()
```

```python
import csv

with open("./data/sample1.csv","r") as f:
    reader = csv.reader(f)
    # 제목 행 제외
    next(reader)

    for c in reader:
        print(c)
```

```python
with open("./data/sample1.csv","r") as f:
    reader = csv.DictReader(f)
    # 제목 행 제외
    next(reader)

    for c in reader:
        print(c)
        for k,v in c.items():
            print(k,v)
```

```python
import json

# 파이썬 : json <=> dict  / 자바스크립트 : json <=> 객체
data = '{"id":"hong", "language":"python", "edition":"3.5","author":"Guido van rossum"}'

print(type(data))

jsondata = json.loads(data)
print(type(jsondata))

print(jsondata['id'])
print(jsondata['language'])
print(jsondata['edition'])
print(jsondata['author'])
```

```python
data = {"id":"hong", "language":"python", "edition":"3.5","author":"Guido van rossum"}

print(type(data))

jsondata = json.dumps(data)
print(type(jsondata))
```

```python
with open("./data/test.json","w") as f:
    json.dump(data,f,indent=2)
```

```python
with open("./data/test.json","r") as f:
    print(json.load(f))
```

- json
    - loads : 문자열을 dict 구조로 처리
    - dumps : dict 구조를 문자열로 처리
    - load : json 파일을 읽어서 dict 구조로 처리
    - dump : dict 구조를 json 으로 처리
