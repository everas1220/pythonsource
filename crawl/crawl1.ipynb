### 크롤링 vs 스크랩핑
- 크롤링 : 주기적으로, 무작위로 데이터 수집하는 것(데이터의 양을 우선으로 수집하는 방식)
    - 사이트에서 허용하는 경로만 가능
- 스크랩핑 : 어떤 특정 카테고리를 선택해서 데이터를 수집(크롤링보다 작은 개념)


```python
import urllib.request as req

url = "http://google.com"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./google.html")
print(header1)
```

```python
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/c/c3/Python-logo-notext.svg/1200px-Python-logo-notext.svg.png"

# urlretrieve
# 요청하는 url 정보를 로컬파일로 저장

file1, header1 = req.urlretrieve(url, "./python.png")
#print(header1)
```

```python
import urllib.request as req

url = "https://encar.com/"

response = req.urlopen(url)
print(response)

print(f"url {response.geturl()}")
print(f"url {response.status}")
print(f"url {response.getheaders()}")

# html 코드 가져오기
contents = response.read().decode("euc-kr")
print(contents[:4000])
```

### requests
- pip install requests
- urllib.request 모듈과 같은 역할

```python
# pip install requests
# pip install fake_useragent

import requests

url = "https://encar.com/"

response = requests.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)
```

```python
# 세션 이용

session = requests.Session()

response = session.get(url)
print(response.status_code)
print(response.headers['content-type'])
print(response.text)

session.close()
```

```python
# get + 파라메터
# ?page=1&size=10

url = "https://httpbin.org/get"

params = {"test":"name"}

res = requests.get(url, params=params)
print(res.headers)
print(res.text)
print(res.json())
```

```python
# post + form data

data = {"name":"hong"}

res = requests.post("https://httpbin.org/post", data=data)
print(res.text)
```

```python
# 행정안전부 rss
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1012
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1013
# https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp?ctxCd=1001

params = []
for code in [1012,1013,1001]:
    params.append(dict(ctxCd=code))

print(params)
url = "https://www.mois.go.kr/gpms/view/jsp/rss/rss.jsp"

with requests.Session() as s:
    for param in params:
        res = s.get(url, params=param)
        print("-"*100)
        print(res.text)
```

```python
from bs4 import BeautifulSoup

with open("./story.html","r") as f:
    html = f.read()

soup = BeautifulSoup(html, "html.parser")
print(soup)
```

```python
# 요소 찾기
# 1. 태그명 이용

print(f"title 요소 찾기 : {soup.title}")
print(f"title 요소의 텍스트 찾기 : {soup.title.text}")
print(f"p 요소 찾기 : {soup.p}")
print(f"b 요소 찾기 : {soup.b}")
```

```python
# 2. 문서의 구조를 이용한 요소 찾기

title = soup.title
print(f"title 의 부모 요소 찾기 : {title.parent}")
children = soup.body.children
for child in children:
    print(f"body 의 자식 요소 : {child}")
```

```python
# 3. find~~~

p1 = soup.p
p2 = p1.find_next_sibling("p")
print(f"첫번째 p 의 다음 형제 요소 : {p2}")
```

```python
# find("태그명","속성") / find_all()

print(f"title : {soup.find("title")}")
print(f"h2 : {soup.find("h2")}")
p = soup.find("p")
print(f"p : {p}")
print(f"p : {p['class']}")
# p2 = soup.find("p",attrs={"class":"story","id":""})
p2 = soup.find("p",class_="title")
print(f"p2 : {p2}, text : {p2.text}")

a1 = soup.a
print(f"a : {a1}, text : {a1.text}")

a2 = soup.find("a", attrs={"class":"sister","id":"link2"})
print(a2)
print(a2.text.strip())

a3 = soup.find("a", id="link3")
print(f"{a3}, text : {a3.text.strip()}")
```

```python
# https://pythonscraping.com/pages/page3.html

url = "https://pythonscraping.com/pages/page3.html"
res = requests.get(url)
soup = BeautifulSoup(res.text,"html.parser")

# Totally Normal Gifts 출력
h1 = soup.find("h1")
print(h1.text)

# 상단 본문 내용 출력
content = soup.select_one('#content')
print(content.text)

# 모든 img 태그 찾기 출력
# src 출력 ['src']
imgs = soup.find_all("img")
for img in imgs:
    print(img)
    print(img['src'])

# 테이블 제목 행
first_row = soup.find_all("th")
for td in first_row:
    print(td)
    print(td.text.strip())

# 선물 가격 출력
cost_rows = soup.select(".gift")
for costs in cost_rows:
    cost = costs.select_one("td:nth-child(3)")
    print(cost.text.strip())
```
