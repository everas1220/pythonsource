### 타이틀 설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html#matplotlib.pyplot.title

```python
x = [1,2,3,4]
y = [2,4,6,8]

plt.title("꺽은선 차트", fontdict={"fontsize": 20})
g = plt.plot(x, y)
```

### 축설정
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html
- https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html

#### 1. 이름 설정
><b>1) xlabel(), ylabel()</b>

```python
plt.title("꺽은선 그래프",fontdict={"size": 20})
#축 이름
plt.xlabel('x 축', fontdict={"color":"red"}, loc='right')
plt.xlabel('y 축',color="#00aa00", loc='top')
g = plt.plot(x, y)
```

><b>2) 이름 및 다양한 설정</b>

matplotlib.pyplot.xlabel(xlabel, fontdict=None, labelpad=None, *, loc=None, **kwargs)

- fontdict : font 설정을 dict 구조로 설정
- labelpad : default 4.0 설정됨(label과 차트 사이의 간격)
- **kwargs
    - color : rgb설정, 색상명 가능
- loc : 축 이름 위치 설정('left', 'center', 'right'),  y 축 (top, bottom, center)

```python
fontdict1 = {"color":"red", "family":"serif", "weight":"bold", 'size':14}
fontdict2 = {"color":"b",  "weight":"bold", 'size':14}


plt.xlabel('x 축', fontdict=fontdict1, loc='right')
plt.xlabel('y 축', fontdict=fontdict2, loc='top')
g = plt.plot(x, y)
```
