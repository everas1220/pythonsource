## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

### 범례 설정
- https://matplotlib.org/stable/api/legend_api.html

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib
x = [1,2,3]
y = [2,4,6]
```

```python
plt.plot(x,y,label="데이터")
g = plt.legend()
```

- 두 개의 선을 하나의 차트에 그릴 때

```python
x1 = [1,2,3]
y1 = [2,4,8]
x2 = [1,2,3]
y2 = [1,4,7]

plt.plot(x1, y1, color ='blue' ,label="data1")
plt.plot(x2, y2, color ='red' ,label="data2")
g = plt.legend()
```

```python
plt.plot(x1, y1, color ='blue')
plt.plot(x2, y2, color ='red')
g = plt.legend(['data1', 'data2'])
```

```python
plt.plot(x, y)
plt.legend(['데이터'],loc='upper center')
```

```python
plt.plot(x, y)
# plt.legend(['데이터'], loc="center center")
g = plt.legend(['데이터'], loc=(0.7,0.3))
```

><b>2) 범례 폰트 사이즈 지정</b>

```python
plt.plot(x1, y1, color ='blue' ,label="data1")
plt.plot(x2, y2, color ='red' ,label="data2")
g = plt.legend(fontsize=20)
```

><b>3) 범례 행 수 지정</b>

```python
plt.plot(x1, y1, color ='blue' ,label="data1")
plt.plot(x2, y2, color ='red' ,label="data2")
g = plt.legend(ncol=2)
```

- [문제] : data1(x : 1, 2, 3; y : 1, 2, 3) 과 data2(x : 1, 2, 3; y : 1, 100, 200) 을 그래프로 출력하시오.

```python
x1 = [1, 2, 3]
y1 = [1, 2, 3]
x2 = [1, 2, 3]
y2 = [1, 100, 200]
```

```python
plt.plot(x1, y1, color='blue')
g = plt.plot(x2, y2, color='red')
```
