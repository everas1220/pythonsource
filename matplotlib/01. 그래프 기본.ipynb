```python
# 라이브러리 로드
import matplotlib.pyplot as plt
import koreanize_matplotlib # 한글 폰트 설정
```

### 기본 차트 생성

><b>리스트를 그래프로 나타내기</b>

```python
x = [1, 2, 3, 4, 5]
y = [12,43,25,15]
g = plt.plot(x, y)
```

><b>파이썬 딕셔너리를 그래프로 나타내기</b>

```python
dict1 = {
    "data_x" : [1,2,3,4,5],
    "data_y" : [2,3,4,10,8]
}

g = plt.plot(dict1["data_x"], dict1["data_y"])
```

### 타이틀 설정

```python
plt.title("꺽은선 차트")
g = plt.plot("data_x", "data_y", data=dict1)
```

```python
x = [1, 2, 3, 4, 5]
y = [12,-5,25,15]

g = plt.plot(x, y)
```

### 그리드

**x축설정**

```python
plt.grid(True, axis='x')
g = plt.plot(x, y)
```
